```python
# Supponiamo che questi siano i punteggi della matrice di pedigree per l'amount
pedigree_01 = [1, 2, 2, 3, 2]  # Affidabilit√†, completezza, ecc.

# Mappa dei punteggi (1-5) ai valori G_i della tabella di ecoinvent
G_map = {1: 0.00, 2: 0.40, 3: 0.80, 4: 1.20, 5: 1.60}

# Converti i punteggi nei valori G_i
G_values = [G_map[score] for score in pedigree_01]

# Standard deviation of the matrix pedigree
sigma_pedigree01 = sqrt(sum(g**2 for g in G_values))
```

```python
# Let's assume these are the pedigree matrix scores for the 'amount'
pedigree_02 = [2, 3, 3, 4, 3]  # Reliability, completeness, etc. 

# Map of scores (1‚Äì5) to the corresponding G_i values from the ecoinvent table
G_map = {1: 0.00, 2: 0.40, 3: 0.80, 4: 1.20, 5: 1.60}

# Convert the scores into their respective G_i values
G_values = [G_map[score] for score in pedigree_02]

# Calculate the standard deviation from the pedigree matrix
sigma_pedigree0102 = sqrt(sum(g**2 for g in G_values))
```

### 26.2.1 Boilers (Natural gas or other blends)

```python
def compute_boiler(lca, indicators):

    modules = ["A13", "A4", "A5", "B1", "B2", "B4", "B6", "C2", "C3", "C4", "D1", "D3"]

    tech = "Boiler"

    for ind in indicators:
        for mod in modules:

            gname = f"lca_{mod}_{tech}_{ind}"

            lname = gname

            if mod == "B1":
                lca[lname] = 0
                continue

            if mod == "B6":
                base  = globals().get(f"lca_B6_Boiler_{ind}", 0)
                extra = globals().get(f"lca_B6_BoilerEle_{ind}", 0)
                lca[lname] = base + extra
                continue

            if mod == "C3":
                lca[lname] = -globals().get(gname, 0)
                continue

            if mod == "C4":
                lca[lname] = -globals().get(gname, 0)
                continue

            lca[lname] = globals().get(gname, 0)

        modD = f"lca_Boiler_ModuleD_{ind}"
        lca[modD] = (
            globals().get(f"lca_D1_Boiler_{ind}", 0) +
            globals().get(f"lca_D3_Boiler_{ind}", 0)
        )

        total_key = f"lca_Boiler_{ind}"
        lca[total_key] = (
            lca.get(f"lca_A13_Boiler_{ind}", 0) +
            lca.get(f"lca_A4_Boiler_{ind}", 0) +
            lca.get(f"lca_A5_Boiler_{ind}", 0) +
            lca.get(f"lca_B2_Boiler_{ind}", 0) +
            lca.get(f"lca_B6_Boiler_{ind}", 0) +
            lca.get(f"lca_C2_Boiler_{ind}", 0) +
            lca.get(f"lca_C3_Boiler_{ind}", 0) +
            lca.get(f"lca_C4_Boiler_{ind}", 0)
        )

    return lca
```

### 26.2.11 Electric Heat pump

```python
def compute_electricheatpump(lca, indicators):

    modules = ["A13", "A4", "A5", "B1", "B2", "B4", "B6", "C2", "C3", "C4", "D1", "D3"]
    tech = "EHP"

    def to_scalar(x):
        if hasattr(x, "score"):
            return float(x)
        elif isinstance(x, (pd.Series, pd.DataFrame)):
            return float(np.nanmean(x.values))
        elif isinstance(x, (np.ndarray, list, tuple)):
            return float(np.nanmean(x))
        elif x is None or (isinstance(x, float) and np.isnan(x)):
            return 0.0
        try:
            return float(x)
        except Exception:
            return 0.0

    def safe_div(a, b):
        a_val, b_val = to_scalar(a), to_scalar(b)
        if b_val == 0:
            return 0.0
        return a_val / b_val

    for ind in indicators:

        for mod in modules:

            gname = f"lca_{mod}_{tech}_{ind}"  
            lname = gname                     

            if mod == "C3":
                lca[lname] = -globals().get(gname, 0)
                continue

            if mod == "C4":
                lca[lname] = -globals().get(gname, 0)
                continue

            if mod == "B6":

                direct_B6 = to_scalar(globals().get(f"lca_B6_{tech}_{ind}", 0))
                B6Ele = to_scalar(globals().get(f"lca_B6_{tech}Ele_{ind}", 0))

                rice_term = safe_div(AFW_RICE, AFQ_RICE) * to_scalar(EHP_EleRICE) * to_scalar(lca.get(f"lca_RICE_{ind}", 0))
                cc_term   = safe_div(AFW_CombinedCycle, AFQ_CombinedCycle) * to_scalar(EHP_EleCombinedCycle) * to_scalar(lca.get(f"lca_CombinedCycle_{ind}", 0))
                orc_term  = safe_div(AFW_ORC, AFQ_ORC) * to_scalar(EHP_EleORC) * to_scalar(lca.get(f"lca_ORC_{ind}", 0))
                gt_term   = safe_div(AFW_GasTurbine, AFQ_GasTurbine) * to_scalar(EHP_EleGasTurbine) * to_scalar(lca.get(f"lca_GasTurbine_{ind}", 0))
                wte_term  = safe_div(AFW_WtE, AFQ_WtE) * to_scalar(EHP_EleWtE) * to_scalar(lca.get(f"lca_WtE_{ind}", 0))

                total_B6 = direct_B6 + B6Ele + rice_term + cc_term + orc_term + gt_term + wte_term

                lca[lname] = total_B6
                continue

            lca[lname] = globals().get(gname, 0)

        lca[f"lca_{tech}_ModuleD_{ind}"] = (
            globals().get(f"lca_D1_{tech}_{ind}", 0) +
            globals().get(f"lca_D3_{tech}_{ind}", 0)
        )

        lca[f"lca_{tech}_{ind}"] = (
            lca.get(f"lca_A13_{tech}_{ind}", 0) +
            lca.get(f"lca_A4_{tech}_{ind}", 0)  +
            lca.get(f"lca_A5_{tech}_{ind}", 0)  +
            lca.get(f"lca_B1_{tech}_{ind}", 0)  +  
            lca.get(f"lca_B2_{tech}_{ind}", 0)  +
            lca.get(f"lca_B6_{tech}_{ind}", 0)  +
            lca.get(f"lca_C2_{tech}_{ind}", 0)  +
            lca.get(f"lca_C3_{tech}_{ind}", 0)  +
            lca.get(f"lca_C4_{tech}_{ind}", 0)
        )

    return lca
```

### 26.2.20 Cooling - Cooling tower

```python
def compute_coolingtower(lca, indicators):

    modules = ["A13", "A4", "A5", "B1", "B2", "B4", "B6", "C2", "C3", "C4", "D1", "D3"]
    tech = "CoolingTower"

    def to_scalar(x):
        if hasattr(x, "score"):
            return float(x)
        elif isinstance(x, (pd.Series, pd.DataFrame)):
            return float(np.nanmean(x.values))
        elif isinstance(x, (np.ndarray, list, tuple)):
            return float(np.nanmean(x))
        elif x is None or (isinstance(x, float) and np.isnan(x)):
            return 0.0
        try:
            return float(x)
        except Exception:
            return 0.0

    def safe_div(a, b):
        a_val, b_val = to_scalar(a), to_scalar(b)
        if b_val == 0:
            return 0.0
        return a_val / b_val

    for ind in indicators:

        for mod in modules:

            gname = f"lca_{mod}_{tech}_{ind}"  
            lname = gname                     

            if mod == "C3":
                lca[lname] = -globals().get(gname, 0)
                continue

            if mod == "C4":
                lca[lname] = -globals().get(gname, 0)
                continue

            if mod == "B6":

                direct_B6 = to_scalar(globals().get(f"lca_B6_{tech}_{ind}", 0))
                B6Ele = to_scalar(globals().get(f"lca_B6_{tech}Ele_{ind}", 0))

                rice_term = safe_div(AFW_RICE, AFQ_RICE) * to_scalar(CoolingTower_EleRICE) * to_scalar(lca.get(f"lca_RICE_{ind}", 0))
                cc_term   = safe_div(AFW_CombinedCycle, AFQ_CombinedCycle) * to_scalar(CoolingTower_EleCombinedCycle) * to_scalar(lca.get(f"lca_CombinedCycle_{ind}", 0))
                orc_term  = safe_div(AFW_ORC, AFQ_ORC) * to_scalar(CoolingTower_EleORC) * to_scalar(lca.get(f"lca_ORC_{ind}", 0))
                gt_term   = safe_div(AFW_GasTurbine, AFQ_GasTurbine) * to_scalar(CoolingTower_EleGasTurbine) * to_scalar(lca.get(f"lca_GasTurbine_{ind}", 0))
                wte_term  = safe_div(AFW_WtE, AFQ_WtE) * to_scalar(CoolingTower_EleWtE) * to_scalar(lca.get(f"lca_WtE_{ind}", 0))

                total_B6 = direct_B6 + B6Ele + rice_term + cc_term + orc_term + gt_term + wte_term

                lca[lname] = total_B6
                continue

            lca[lname] = globals().get(gname, 0)

        lca[f"lca_{tech}_ModuleD_{ind}"] = (
            globals().get(f"lca_D1_{tech}_{ind}", 0) +
            globals().get(f"lca_D3_{tech}_{ind}", 0)
        )

        lca[f"lca_{tech}_{ind}"] = (
            lca.get(f"lca_A13_{tech}_{ind}", 0) +
            lca.get(f"lca_A4_{tech}_{ind}", 0)  +
            lca.get(f"lca_A5_{tech}_{ind}", 0)  +
            lca.get(f"lca_B1_{tech}_{ind}", 0)  +  
            lca.get(f"lca_B2_{tech}_{ind}", 0)  +
            lca.get(f"lca_B6_{tech}_{ind}", 0)  +
            lca.get(f"lca_C2_{tech}_{ind}", 0)  +
            lca.get(f"lca_C3_{tech}_{ind}", 0)  +
            lca.get(f"lca_C4_{tech}_{ind}", 0)
        )

    return lca
```

### 26.2.21 Auxiliary

```python
def compute_auxiliary(lca, indicators):

    modules = ["A13", "A4", "A5", "B1", "B2", "B4", "B6", "C2", "C3", "C4", "D1", "D3"]
    tech = "Auxiliary"

    for ind in indicators:
        for mod in modules:

            gname = f"lca_{mod}_{tech}_{ind}"
            lname = gname  # chiave identica nel dizionario finale

            if mod == "A5":
                lca[lname] = globals().get(f"lca_A5_Pipes_{ind}", 0)
                continue

            if mod == "C3":
                lca[lname] = -globals().get(gname, 0)
                continue

            if mod == "C4":
                lca[lname] = -globals().get(gname, 0)
                continue

            if mod == "B6":
                lca[lname] = globals().get(gname, 0)
                continue

            if mod == "B1":
                lca[lname] = globals().get(gname, 0)
                continue

            lca[lname] = globals().get(gname, 0)

        lca[f"lca_{tech}_ModuleD_{ind}"] = (
            globals().get(f"lca_D1_{tech}_{ind}", 0) +
            globals().get(f"lca_D3_{tech}_{ind}", 0)
        )

        lca[f"lca_{tech}_{ind}"] = (
            lca.get(f"lca_A13_{tech}_{ind}", 0) +
            lca.get(f"lca_A4_{tech}_{ind}", 0)  +
            lca.get(f"lca_A5_{tech}_{ind}", 0)  +
            lca.get(f"lca_B1_{tech}_{ind}", 0)  +
            lca.get(f"lca_B2_{tech}_{ind}", 0)  +
            lca.get(f"lca_B6_{tech}_{ind}", 0)  +
            lca.get(f"lca_C2_{tech}_{ind}", 0)  +
            lca.get(f"lca_C3_{tech}_{ind}", 0)  +
            lca.get(f"lca_C4_{tech}_{ind}", 0)
        )

    return lca
```

### 26.2.22 - Sum per module and total

```python
def compute_sum(lca, indicators):

    techs = [
        "Biomassboiler", "CombinedCycle", "Fueloilboiler", "GasTurbine",
        "HeatRecovery", "Boiler", "ORC", "RICE", "SolarThermal",
        "ThermalConcreteStorage", "ThermalSteelStorage", "WtE",
        "EHP", "GAHP", "GEHP", "GEO", "AbsChiller", "CompressionChiller", 
        "DryCooler", "CoolingTower", "Auxiliary", 
    ]

    modules = ["A13", "A4", "A5", "B1", "B2", "B6", "C2", "C3", "C4"]

    for ind in indicators:

        for mod in modules:
            total = 0
            for tech in techs:
                key = f"lca_{mod}_{tech}_{ind}"
                total += lca.get(key, 0)
            lca[f"lca_{mod}_{ind}"] = total

        # --- MODULE D ---
        moduleD_total = 0
        for tech in techs:
            key = f"lca_{tech}_ModuleD_{ind}"
            moduleD_total += lca.get(key, 0)
        lca[f"lca_ModuleD_{ind}"] = moduleD_total

        lca[f"lca_totale_{ind}"] = (
            lca.get(f"lca_A13_{ind}", 0)
            + lca.get(f"lca_A4_{ind}", 0)
            + lca.get(f"lca_A5_{ind}", 0)
            + lca.get(f"lca_B2_{ind}", 0)
            + lca.get(f"lca_B6_{ind}", 0)
            + lca.get(f"lca_C2_{ind}", 0)
            + lca.get(f"lca_C3_{ind}", 0)
            + lca.get(f"lca_C4_{ind}", 0)
        )

    return lca
```

```python
lca = {}

zero_techs = [
    "CombinedCycle",
    "GasTurbine",
    "ORC",
    "RICE",
    "WtE",
]

LCA_MODULES = ["A13", "A4", "A5", "B1", "B2", "B4", "B6", "C2", "C3", "C4", "D"]

for ind in indicators:

    if DHTypology == "5th generation":
        denom = HeatingEnergyDHC + CoolingEnergyDHC

        if denom == 0:
            HeatingEnergyDH = 0.0
        else:
            if Aim == "Heating":
                HeatingEnergyDH = HeatingEnergyDHC / denom
            elif Aim == "Cooling":
                HeatingEnergyDH = CoolingEnergyDHC / denom
            else:
                HeatingEnergyDH = 1.0
    else:
        HeatingEnergyDH = 1.0

    compute_boiler(lca, [ind])
    compute_biomassboiler(lca, [ind])
    compute_fueloilboiler(lca, [ind])

    # üîß FIX PER COOLING 3rd / 4th GEN
    if Aim == "Cooling" and DHTypology in ["3rd generation", "4th generation"]:
        for tech in zero_techs:
            for mod in LCA_MODULES:
                var_name = f"lca_{tech}_{mod}_{ind}"
                globals()[var_name] = 0.0
    else:
        compute_combinedcycle(lca, [ind])
        compute_turbogas(lca, [ind])
        compute_orc(lca, [ind])
        compute_rice(lca, [ind])
        compute_wte(lca, [ind])

    compute_electricheatpump(lca, [ind])
    compute_gasabsorptionheatpump(lca, [ind])
    compute_gasengineheatpump(lca, [ind])
    compute_geothermal(lca, [ind])
    compute_solarthermal(lca, [ind])
    compute_thermalconcretestorage(lca, [ind])
    compute_thermalsteelstorage(lca, [ind])
    compute_heatrecovery(lca, [ind])
    compute_absorptionchiller(lca, [ind])
    compute_compressionchiller(lca, [ind])
    compute_drycooler(lca, [ind])
    compute_coolingtower(lca, [ind])
    compute_auxiliary(lca, [ind])

    if DHTypology == "5th generation":
        for tech in lca:
            if ind in lca[tech]:
                lca[tech][ind] *= HeatingEnergyDH
```
